```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/medical_appointment_data.csv")

df.info()
df.describe()

# Target distribution
sns.countplot(x="no_show", data=df)
plt.title("No-Show Distribution")
plt.show()

# Age vs No-show
sns.boxplot(x="no_show", y="age", data=df)
plt.show()

# SMS vs No-show
sns.barplot(
    x="SMS_received",
    y=df["no_show"].eq("Yes").astype(int),
    data=df
    
)
plt.show()
```
